```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# sentences
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
# define the vocab size 
voc_size = 10000
```

```python
one_hot_repr = [one_hot(words,voc_size) for words in sent]
```

```python
one_hot_repr
```

the output above shows index number where these words will be present in vocab,
if we expand each word all other index will be zero and only this index will be one

```python
#word embedding representation
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

all sentences must be of same length, so that is why we use pad_sequence
The vectors which are not of 8 length, 0s will be added to make them 8 characters long


```python
#setting max sentence length 
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
#padding = pre means 0s will be added in start to make vector length = 8
# we can also use 'post'
```

```python
print(embedded_docs)
```

```python
# feature representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.build(input_shape=(None, sent_length))
model.compile('adam','mse') #mse is loss function
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
